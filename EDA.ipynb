```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
historico_vel=pd.read_csv("historico_velocidad_internet.csv",low_memory=False,decimal=',')
```

```python
historico_vel.head(5)
```

```python
group_historico = historico_vel.groupby("Año")["Mbps (Media de bajada)"].mean()
```

```python
#Evolución de la velocidad a lo largo del tiempo
group_historico.plot(kind='bar')
plt.xlabel("Año")
plt.ylabel("Media de bajada")
plt.title("Mbps x año ")
plt.show()
```

```python
ingresos=pd.read_csv("Internet_Ingresos.csv",low_memory=False,thousands='.')
```

```python
ingresos.head(5)
```

```python
ingresos["Ingresos (miles de pesos)"]=pd.to_numeric(["Ingresos (miles de pesos)"])
```

```python
group_ingresos = ingresos.groupby("Año")["Ingresos (miles de pesos)"].mean()
```

```python
#Evolución de los ingresos a lo largo del tiempo
group_ingresos.plot(kind='bar')
plt.xlabel("Año")
plt.ylabel("Ingresos")
plt.title("Ingresos x año ")
plt.show()
```

```python
penetracion=pd.read_csv("Internet_Penetracion.csv",low_memory=False,decimal=',')
```

```python
penetracion=penetracion[penetracion.Año == 2022]
penetracion=penetracion[penetracion.Trimestre == 3]
penetracion=penetracion.sort_values("Accesos por cada 100 hogares")
```

```python
penetracion.head(5)
```

```python
baf=pd.read_csv("Internet_BAF.csv",low_memory=False,thousands='.')
dial_up = baf.groupby("Año")["Dial up"].sum()
Banda_ancha=baf.groupby("Año")["Banda ancha fija"].sum()
```

```python
baf.head(5)
```

```python
#Evolución del Dial-Up
dial_up.plot(kind='bar')
plt.xlabel("Año")
plt.ylabel("Ingresos")
plt.title("Evolución de Dial-Up")
plt.show()
```

```python
#Evolución de la banda ancha
Banda_ancha.plot(kind='bar')
plt.xlabel("Año")
plt.ylabel("Ingresos")
plt.title("Evolución de Banda Ancha")
plt.show()
```
